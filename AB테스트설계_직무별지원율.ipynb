# 이탈률 분석 및 개선 전략

## 1. 북마크 여부에 따른 이탈률 분석
### 가설
- 북마크가 있는 공고의 이탈률이 낮을 것이다.

### 분석 방법
- 북마크 유무에 따른 이탈률 비교.

### 조치 방안
- 북마크 수를 증가시키기 위해 다음을 추진:
  - 북마크 기능 개선(UX/UI 개선).
  - 북마크 유도 마케팅 캠페인 진행.

---

## 2. 기업 팔로우 여부에 따른 이탈률 분석
### 가설
- 기업 팔로우 수가 이탈률에 영향을 미칠 것이다.

### 분석 방법
- 기업 팔로우 유무에 따른 이탈률 비교.

### 조치 방안
- 기업 팔로우를 활성화하기 위해:
  - 팔로우 유도 메시지 및 알림 추가.
  - 팔로우 혜택(예: 공고 알림, 전용 콘텐츠 등) 제공.

---

## 3. 직무 카테고리별 이탈률 분석
### 가설
- 특정 직무 카테고리가 높은 이탈률을 보일 것이다.

### 분석 방법
- 직무 카테고리별 이탈률 분석.

### 조치 방안
- 이탈률이 높은 직무에 대한 맞춤형 개선 전략:
  - 해당 직무 공고에 매력적인 콘텐츠 추가.
  - 사용자 친화적인 설명 및 혜택 강조.

---

## 4. 검색 필터 사용 패턴에 따른 이탈률 분석
### 가설
- 특정 검색 필터를 많이 사용하는 유저의 이탈률이 높을 것이다.

### 분석 방법
- 검색 필터별 이탈률 분석.

### 조치 방안
- 자주 사용되는 필터의 사용성을 개선하거나 추천 기능 도입:
  - 검색 필터 추천 기능 추가.
  - 검색 필터의 UI 개선.

---

## 5. 공고페이지 체류 시간에 따른 이탈률 분석
### 가설
- 공고페이지 체류 시간이 길어질수록 이탈률이 높을 것이다.

### 분석 방법
- **체류 시간 계산**: 공고페이지 접속시간 - 지원 시작시간.
  - 단, 이탈한 경우 지원 시작시간 데이터를 알 수 없음을 고려.

### 조치 방안
- 체류 시간을 줄이기 위해:
  - 주요 정보를 페이지 상단에 배치.
  - 공고 요약 기능 추가.

---

## 6. 코호트 분석
### 목표
- 사용자 그룹을 날짜별로 분류하여 이탈률이 높은 시기와 낮은 시기를 식별.

### 분석 방법
- 사용자 ID별로 코호트를 나누고 특정 시기의 이탈률 비교.

### 조치 방안
- 이탈률이 높은 시기:
  - 원인 파악 및 개선 조치(예: 기능 장애, UI 개선).
- 이탈률이 낮은 시기:
  - 해당 시기의 긍정적 요인을 강화.

### 쿼리와 path로 세분화해서 나누기

```python
from urllib.parse import parse_qs, urlparse

# 쿼리 파싱 함수
def parse_query(path):
    query = urlparse(path).query
    return parse_qs(query)

# 경로별 쿼리 분석
parsed_queries = jobs2_df['이벤트 발생 화면 경로'].apply(parse_query)
expanded_df = pd.json_normalize(parsed_queries)

result = pd.concat([jobs2_df, expanded_df], axis=1)
```

## 데이터 변수 설명
1. **job**: 직무 카테고리
   - 숫자로 직무 형태를 나타냄.
   - 예: `1=개발자`, `2=디자이너` 등.
   - 사용자가 선택한 직무를 필터링할 가능성이 높음.

2. **page**: 페이지 번호

3. **salary**: 연봉 설정
   - 사용자가 연봉을 기준으로 필터링한 값을 나타냄.

4. **location**: 근무 지역
   - 사용자가 근무 지역으로 필터링한 값을 나타냄.

5. **q**: 쿼리
   - 사용자의 검색 입력과 관련된 정보를 나타냄. (자세한 정보는 추가 분석 필요)

6. **keywords**: 검색어
   - 사용자가 입력한 검색어.

7. **specialty**: 활동 분야
   - 사용자의 특정 전문 분야나 특화된 활동을 나타냄.

8. **sort**: 정렬 기준
   - 사용자가 공고를 정렬한 기준.
   - 예: 최신순, 연봉순 등.

9. **career_type**: 경력 형태
   - 사용자가 필터링한 경력 조건.

```python
공통된_지원자['직무'].value_counts()
```

```python
# 공통된_지원자 데이터의 직무별 지원자 수 계산
직무별_지원자수 = 공통된_지원자['직무'].value_counts()
print(직무별_지원자수)
# 채용공고방문USER의 전체 유저 ID 개수 계산
전체_유저_수 = 채용공고방문자['유저 ID'].nunique()  # 중복 제거한 유저 수
print(전체_유저_수)
# 직무별 지원율 계산
직무별_지원율 = (직무별_지원자수 / 전체_유저_수) * 100  # 비율을 퍼센트로 변환

# 지원율 결과 출력
print(직무별_지원율)
```

```python
# 공통된_지원자 데이터의 직무별 지원자 수 계산
직무별_지원자수 = 공통된_지원자['직무'].value_counts()

# 채용공고방문자의 전체 유저 ID 개수 계산
전체_유저_수 = 채용공고방문자['유저 ID'].nunique()

# 직무별 지원율 계산
직무별_지원율 = (직무별_지원자수 / 전체_유저_수) * 100

# 직무별 지원자수와 지원율을 합쳐 DataFrame 생성
결과_데이터_2022 = pd.DataFrame({
    '직무': 직무별_지원자수.index,          # 직무 이름
    '지원자수': 직무별_지원자수.values,    # 지원자 수
    '지원율': 직무별_지원율.values         # 지원율
})
```

```python
# '지원서 제출시간' 컬럼을 datetime 형식으로 변환
지원자['지원서 제출시간'] = pd.to_datetime(지원자['지원서 제출시간'], errors='coerce')

# 2022년 데이터 필터링
지원자_2023 = 지원자[지원자['지원서 제출시간'].dt.year == 2023]

# 결과 확인
print(지원자_2023)
```

```python
지원자_2023.drop_duplicates(subset='유저 ID', keep='first', inplace=True)
```

```python
채용공고방문자_2023 = pd.read_csv('/Users/parkjunhyeok/Desktop/codit/프로젝트/중급프로젝트/전처리data/채용공고방문_2023.csv',encoding='utf-8-sig')
```

```python
공통된_지원자 = pd.merge(
    채용공고방문자_2023,
    지원자_2023,
    on='유저 ID',
    how='inner'  # 교집합으로 병합
)
```

```python
공통된_지원자['직무'].value_counts()
```

```python
# 공통된_지원자 데이터의 직무별 지원자 수 계산
직무별_지원자수 = 공통된_지원자['직무'].value_counts()

# 채용공고방문자의 전체 유저 ID 개수 계산
전체_유저_수 = 채용공고방문자['유저 ID'].nunique()

# 직무별 지원율 계산
직무별_지원율 = (직무별_지원자수 / 전체_유저_수) * 100

# 직무별 지원자수와 지원율을 합쳐 DataFrame 생성
결과_데이터_2023 = pd.DataFrame({
    '직무': 직무별_지원자수.index,          # 직무 이름
    '지원자수': 직무별_지원자수.values,    # 지원자 수
    '지원율': 직무별_지원율.values         # 지원율
})
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.rc('font', family='AppleGothic')

top_jobs = 결과_데이터_2023.head(10)

plt.figure(figsize=(12, 8))
sns.barplot(data=top_jobs, x='지원율', y='직무', palette='viridis')
plt.xlabel('지원율 (%)')
plt.ylabel('직무')
plt.title('2023년도 직무별 지원율 상위 10개')
plt.show()
```
